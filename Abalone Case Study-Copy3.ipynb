# Abalone Case Study

The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope -- a boring and time-consuming task. Other measurements, which are easier to obtain, are used to predict the age.

# Attribute Information

Given is the attribute name, attribute type, the measurement unit and a brief description. The number of rings is the value to predict.   

Name / Data Type / Measurement Unit / Description  
-----------------------------
Sex / nominal / -- / M, F, and I (infant)  
Length / continuous / mm / Longest shell measurement  
Diameter / continuous / mm / perpendicular to length  
Height / continuous / mm / with meat in shell  
Whole weight / continuous / grams / whole abalone  
Shucked weight / continuous / grams / weight of meat  
Viscera weight / continuous / grams / gut weight (after bleeding)  
Shell weight / continuous / grams / after being dried  
Rings / integer / -- / +1.5 gives the age in years.

# Model Building

```python
# dividing the dataset into independent and dependent variables
x=df_new_aba.drop("Rings", axis=1)
y=df_new_aba["Rings"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
# Predicting the value
pred_lr=lr.predict(x_test)
print("Predicted value: ",pred_lr)
print("Actual value: ",y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

print('Error:')

print('Mean absolute error: ',mean_absolute_error(y_test,pred_lr))
print('Mean squared error: ',mean_squared_error(y_test,pred_lr))

print('Root Mean squared error: ',np.sqrt(mean_squared_error(y_test,pred_lr)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred_lr))
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(lr,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

# Random Forest Regressor

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters = {'criterion':['mse','mae'], 'max_features':["auto", "sqrt"]}
rfr =RandomForestRegressor()
gd= GridSearchCV(rfr,parameters)
gd.fit(x_train,y_train)

print(gd.best_params_)
```

```python
rf= RandomForestRegressor(criterion="mae",max_features="sqrt")
rf.fit(x_train, y_train)
rf.score(x_train, y_train)
```

```python
pred_rfr = rf.predict(x_test)
print("Predicted value: ",pred_rfr)
print("Actual value: ",y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

print('Error:')

print('Mean absolute error: ',mean_absolute_error(y_test,pred_rfr))
print('Mean squared error: ',mean_squared_error(y_test,pred_rfr))

print('Root Mean squared error: ',np.sqrt(mean_squared_error(y_test,pred_rfr)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred_rfr))
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(rfr,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

From the above models, Lasso ,Ridge and Linear Regression show similar r2 scores and cross validation scores. 

The highest cross validation and r2 score belong to the Random Forest Regressor model. Also, the lowest cross validation score and r2 score belong to Decision tree Regressor model.
